```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%inline matplotlib
```

```python
df=sns.load_dataset("titanic")
```

```python
df
```

```python
df.columns
```

```python
plt.figure(figsize=(6,6))
df['sibsp'].value_counts().plot(kind='bar')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(7,7))
df['pclass'].value_counts().plot(kind='pie')
plt.legend()
plt.show()
```

```python
df.pclass.describe()
```

```python
#Univariate
```

```python
sns.stripplot(y='sibsp',data=df,jitter=True)
plt.show()
```

```python
sns.stripplot(y='sibsp',data=df,jitter=False)
plt.show()
```

```python
sns.stripplot(y='pclass',data=df)
plt.show()
```

```python
# Bivariate
```

```python
plt.figure(figsize=(6,6))
sns.stripplot(x='pclass',y='fare',data=df)
plt.ylabel('fare')
plt.show()
```

```python
sns.boxplot(x='pclass',y='fare',data=df)
plt.xlabel('pclass',size=15)
plt.ylabel('fare',size=15)
plt.show()
```

```python
sns.stripplot(x='sex',y='fare',data=df)
plt.xlabel('sex',size=15)
plt.ylabel('fare',size=15)
plt.show()
```

```python
sns.stripplot(x='sex',y='fare',data=df)
plt.xlabel('pclass',size=15)
plt.ylabel('fare',size=15)
plt.show()
```

```python
sns.stripplot(x="pclass",y="fare",hue="sex",data=df)
plt.xlabel("pclass",size=15)
plt.ylabel("fare",size=15)
plt.show()
```
